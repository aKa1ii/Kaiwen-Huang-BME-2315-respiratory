### 1. Fitting the SIR model

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# euler_sir is not exported from main_functions in your environment,
# so import only the available helper and define a local euler_sir here.
from main_functions import convert_cumulative_to_SIR


def euler_sir(beta, gamma, S0, I0, R0, t_array, N):
    """
    Simple forward Euler solver for the SIR equations.
    t_array: array-like of time points (can be non-uniform); returns arrays same length.
    """
    t = np.asarray(t_array, dtype=float)
    n = len(t)
    S = np.zeros(n, dtype=float)
    I = np.zeros(n, dtype=float)
    R = np.zeros(n, dtype=float)
    S[0], I[0], R[0] = float(S0), float(I0), float(R0)

    for k in range(1, n):
        dt = t[k] - t[k - 1]
        # derivatives at previous step
        dS = -beta * S[k - 1] * I[k - 1] / N
        dI = beta * S[k - 1] * I[k - 1] / N - gamma * I[k - 1]
        dR = gamma * I[k - 1]

        S[k] = S[k - 1] + dS * dt
        I[k] = I[k - 1] + dI * dt
        R[k] = R[k - 1] + dR * dt

    return S, I, R


data = pd.read_csv('/Users/mollykessenich/Desktop/BME 2315/Module_4/Code/ebola_sierra_leone_data_2014_2015_cumulative.csv', parse_dates=['date'])

data['cumulative_cases'] = data['confirmed_cases']

data_sir = convert_cumulative_to_SIR(
    data,
    date_col='date',
    cumulative_col='cumulative_cases',
    population=7075641
)

I_obs = data_sir['I_est'].values.astype(float)
t_obs = np.linspace(0, len(I_obs) - 1, len(I_obs))  # time array in days

I0_obs = float(data_sir.iloc[0]['I_est'])
R0_obs = float(data_sir.iloc[0]['R_est'])

N = float(data_sir.loc[0, 'S_est'] + data_sir.loc[0, 'I_est'] + data_sir.loc[0, 'R_est'])

S0_obs = N - I0_obs - R0_obs

beta1 = 0.025
gamma1 = 1 / 48
beta2 = 0.02
gamma2 = 1 / 55

S1, I1, R1 = euler_sir(beta1, gamma1, S0_obs, I0_obs, R0_obs, t_obs, N)
S2, I2, R2 = euler_sir(beta2, gamma2, S0_obs, I0_obs, R0_obs, t_obs, N)

plt.plot(t_obs, I1, label=f'beta={beta1},gamma={gamma1}', marker='x')
plt.plot(t_obs, I2, label=f'beta={beta2},gamma={gamma2}', marker='s')
plt.plot(t_obs, I_obs, 'o', label='Observed')

plt.legend()
plt.xlabel('days')
plt.ylabel('Infections')
plt.title('Demo: effect of beta/gamma on I(t)')
plt.show()

print('MSE for params 1:', np.mean((I1 - I_obs) ** 2))
print('MSE for params 2:', np.mean((I2 - I_obs) ** 2))
```

```python

from scipy.optimize import minimize

def sse_from_params(params, S0, I0, R0, t_obs, N, I_obs):
    beta, gamma = params

    if beta <= 0 or gamma <= 0:
        return np.inf

    S_model, I_model, R_model = euler_sir(beta, gamma, S0, I0, R0, t_obs, N)
    sse = np.sum((I_model - I_obs) ** 2)
    return sse

initial_guess = [0.8, 0.9]


bounds = [(1e-6, 1.0),  
          (1/60, 1.0)]   

result = minimize(
    sse_from_params,
    x0=initial_guess,
    args=(S0_obs, I0_obs, R0_obs, t_obs, N, I_obs),
    bounds=bounds
)

beta_opt, gamma_opt = result.x

print(f"Optimal beta  = {beta_opt:.6f}")
print(f"Optimal gamma = {gamma_opt:.6f}")
print(f"Minimum SSE   = {result.fun:.6e}")

S_opt, I_opt, R_opt = euler_sir(beta_opt, gamma_opt, S0_obs, I0_obs, R0_obs, t_obs, N)

plt.figure()
plt.plot(t_obs, I_obs, 'o', label='Observed I(t)')
plt.plot(t_obs, I_opt, '-', label=f'Best fit I(t)\nβ={beta_opt:.4f}, γ={gamma_opt:.4f}')
plt.xlabel('Days')
plt.ylabel('Infections (in millions)')
plt.title('Best-fitting SIR model')
plt.legend()
plt.show()
```

**Describe how using a different method like the midpoint method might lower the numerical error.**
Using a different numerical method like the midpoint method can lower numerical error because it uses additional information about how the system is changing within each time step. Euler’s method estimates the next value based only on the slope at the beginning of the step, which can cause the solution to drift away from the true trajectory, especially when the step size is large or the system is changing quickly. In contrast, the midpoint method takes an extra step to estimate the slope halfway through the interval. This midpoint slope is usually a better approximation of the average rate of change over the entire step, which reduces the accumulated error. As a result, the midpoint method typically follows the true solution more closely than Euler’s method, even when using the same step size.

#### 3. Decreasing numerical error with the RK4 Method

```python
from scipy.integrate import solve_ivp

def sir_ode(t, y, beta, gamma, N):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]

def run_sir_ivp(beta, gamma, S0, I0, R0, t_eval, N):
    y0 = [S0, I0, R0]
    sol = solve_ivp(
        fun=lambda t, y: sir_ode(t, y, beta, gamma, N),
        t_span=(t_eval[0], t_eval[-1]),
        y0=y0,
        t_eval=t_eval,     
        method='RK45'       
    )
    S, I, R = sol.y
    return S, I, R

n = len(I_obs)
split_idx = n // 2
I_first = I_obs[:split_idx]

def sse_first_half_rk4(params):
    beta, gamma = params
    if beta <= 0 or gamma <= 0:
        return np.inf

    S_rk, I_rk, R_rk = run_sir_ivp(beta, gamma, S0_obs, I0_obs, R0_obs, t_obs, N)
    I_rk_first = I_rk[:split_idx]
    return np.sum((I_rk_first - I_first) ** 2)

initial_guess = [0.05, 1/15]
bounds = [(1e-6, 1.0), (1/60, 1.0)]

result_rk = minimize(
    sse_first_half_rk4,
    x0=initial_guess,
    bounds=bounds
)

beta_rk, gamma_rk = result_rk.x
print(f"RK4 best-fit on FIRST half: beta = {beta_rk:.6f}, gamma = {gamma_rk:.6f}")
print(f"Min SSE (first half, RK4): {result_rk.fun:.6e}")

S_rk_all, I_rk_all, R_rk_all = run_sir_ivp(beta_rk, gamma_rk,
                                           S0_obs, I0_obs, R0_obs,
                                           t_obs, N)


sse_second_half_rk4 = np.sum((I_rk_all[split_idx:] - I_obs[split_idx:]) ** 2)
print(f"SSE on SECOND half (RK4): {sse_second_half_rk4:.6e}")

plt.figure()
plt.plot(t_obs, I_obs, 'o', label='Observed I(t)')
plt.plot(t_obs, I_rk_all, '-', label=f'RK4 fit from first half (β={beta_rk:.4f}, γ={gamma_rk:.4f})')
plt.axvline(t_obs[split_idx-1], color='gray', linestyle='--', label='Train / future split')
plt.xlabel('Days')
plt.ylabel('Infections (in millions)')
plt.title('SIR (RK4 / solve_ivp) fit on FIRST half, prediction on SECOND half')
plt.legend()
plt.show()
```

```python
# Extended model implementation - including an exposed compartment (SEIR model)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import minimize

# Make sure observed daily new cases exist
#    (I_obs should already have been built earlier from cumulative data)
I_obs = I_obs.astype(float)

# Time array (days since start)
t_obs = np.arange(len(I_obs))

# Population and initial conditions
N = 7075641
I0_obs = float(I_obs[0])
R0_obs = 0.0
S0_obs = N - I0_obs - R0_obs

# 1. Train / Test split (first vs. second half)
split_idx = len(t_obs) // 2

t_first  = t_obs[:split_idx]
t_second = t_obs[split_idx:]

I_first  = I_obs_smooth[:split_idx]
I_second = I_obs_smooth[split_idx:]

S0_first = S0_obs
I0_first = float(I_obs_smooth[0])
R0_first = 0.0

# SEIR needs an initial exposed value - assume E0 similar to I0:
E0_first = I0_first 

# SEIR ODE system
def seir_ode(t, y, beta, sigma, gamma, N):
    S, E, I, R = y
    dS = -beta * S * I / N
    dE = beta * S * I / N - sigma * E
    dI = sigma * E - gamma * I
    dR = gamma * I
    return [dS, dE, dI, dR]

# RK4 solver wrapper (solve_ivp RK45)
def seir_rk4(beta, sigma, gamma, S0, E0, I0, R0, t_array, N):
    sol = solve_ivp(
        lambda t, y: seir_ode(t, y, beta, sigma, gamma, N),
        (t_array[0], t_array[-1]),
        [S0, E0, I0, R0],
        t_eval=t_array,
        method="RK45"
    )
    S, E, I, R = sol.y
    return S, E, I, R

# Fit incidence (daily new infections), not I(t)
#    Incidence(t) = beta * S(t) * I(t) / N
def seir_sse_incidence(params, S0, E0, I0, R0, t_array, I_data_daily, N):
    beta, sigma, gamma = params
    if beta <= 0 or sigma <= 0 or gamma <= 0:
        return np.inf

    S, E, I, R = seir_rk4(beta, sigma, gamma, S0, E0, I0, R0, t_array, N)
    incidence_pred = beta * S * I / N  # predicted daily new cases

    return np.sum((incidence_pred - I_data_daily) ** 2)

# Fit SEIR on first half
initial_guess = [0.2, 1/7, 0.1]                 # beta, sigma, gamma
bounds = [(1e-6, 2.0), (1e-6, 1.0), (1e-6, 1.0)]

res_seir = minimize(
    seir_sse_incidence,
    initial_guess,
    args=(S0_first, E0_first, I0_first, R0_first, t_first, I_first, N),
    bounds=bounds
)

beta_seir, sigma_seir, gamma_seir = res_seir.x # sigma models the latent period
print(f"SEIR best-fit (first half): beta={beta_seir:.6f}, sigma={sigma_seir:.6f}, gamma={gamma_seir:.6f}")
print("SEIR fit success:", res_seir.success, "-", res_seir.message)

# Predict full trajectory and compute incidence
S_seir, E_seir, I_seir, R_seir = seir_rk4(
    beta_seir, sigma_seir, gamma_seir,
    S0_first, E0_first, I0_first, R0_first,
    t_obs, N
)

incidence_full = beta_seir * S_seir * I_seir / N  # predicted daily new cases for whole time window

# SSE on second half (incidence vs smoothed observed daily new cases)
sse_seir_second = np.sum((incidence_full[split_idx:] - I_obs_smooth[split_idx:]) ** 2)
print(f"SEIR SSE on second half (incidence): {sse_seir_second:.6e}")

# Plot observed vs predicted daily new cases

plt.figure(figsize=(9,5))
plt.plot(t_obs, I_obs, 'o', alpha=0.35, label='Observed daily new cases (raw)')
plt.plot(t_obs, I_obs_smooth, '-', label='Observed daily new cases (7-day avg)')
plt.plot(t_obs, incidence_full, '-', linewidth=2, label='SEIR predicted daily new cases')
plt.axvline(split_idx, color='k', linestyle='--', label='Train/Test split')
plt.xlabel('days since start')
plt.ylabel('New cases per day')
plt.title('SEIR model fit (incidence) vs observed daily new cases')
plt.grid(True)
plt.legend()
plt.show()

```

Based on the results of the SEIR model fit, there was a high transmission rate (beta was about 0.99) and unrealistically fast incubation and recovery rates (sigma = gamma = 1 day). This indicates the optimizer pushed parameters to unrealistically high values to try and explain the sharp early peak in cases. While the SEIR model captures the overall declining trend of the outbreak, it fails to reproduce the rapid rise and fall and the variability in daily reported cases. The SSE on the second half of the data was not reduced significantly compared to the SIR model, suggesting that the dominant source of error arises from model assumptions such as constant parameters, homogeneous mixing, and unmodeled intervention effects, rather than from missing compartments.
